```python
import bs4
import requests

request = requests.get('https://www.dba.dk/', headers={'User-Agent': 'Mozilla/5.0'})
request.raise_for_status()
soup = bs4.BeautifulSoup(request.text, "html.parser")

print(soup)
```

```python
kategorier = soup.select("li.single > a")

for e in kategorier:
    print(e.get("title"))
```

```python
links = soup.select("a")

for a in links:
    if not str(a.get('href')).startswith('https'):
        continue
    print(a.get('href'))
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options

options = Options()
options.headless = True
browser = webdriver.Firefox(options=options)

base_url = "https://www.dba.dk/"
browser.get(base_url)
browser.implicitly_wait(3)

biler_link = browser.find_elements_by_link_text("Biler")
biler_link[0].click()

print(biler_link)
```

```python
import matplotlib.pyplot as plt

request = requests.get('https://www.dba.dk/biler/', headers={'User-Agent': 'Mozilla/5.0'})
request.raise_for_status()
soup = bs4.BeautifulSoup(request.text, "html.parser")

car_array = []
area_array = []

car_soup = soup.select("nav.row-fluid:nth-child(3) > ul:nth-child(2) > li > a")

for area in car_soup:
    area_array.append(area.text.split(" (")[0])
    
for sales in car_soup:
    car_array.append(sales.text.split("(")[1].split(")")[0])
    
print(area_array)
print(car_array)

plt.bar(area_array, car_array,width=0.5, align='center')
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
```
