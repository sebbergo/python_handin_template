  1. `def print_file_content(file)` that can print content of a csv file to the console
  2. `def write_list_to_file(output_file, lst)` that can take a list of tuple and write each element to a new line in file

  3. `def read_csv(input_file)` that take a csv file and read each row into a list

```python
import csv

file_name = './exercise1.csv'

def print_file_content(imported_file):
    with open(imported_file, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)

print_file_content(file_name)
```

```python
import csv

file_name = './exercise2.csv'

my_lst = ("hej", "Seb", "går", "det", "godt")

def write_lst_to_file(imported_file, lst):
    with open(imported_file, 'w') as file:
        write = csv.writer(file)
        for element in lst:
            file.write(element + "\n")
            print(element)

write_lst_to_file(file_name, my_lst)
```

```python
import csv

file_name = './exercise2.csv'

string1 = "Hej"
string2 = "Seb"
string3 = "går"
string4 = "det"
string5 = "godt"

def write_lst_to_file(imported_file, *strings):
    with open(imported_file, 'w') as file:
        write = csv.writer(file)
        for string in strings:
            file.write(string + "\n")
            print(string)

write_lst_to_file(file_name, string1, string2, string3,
                 string4, string5)
```

```python
import csv

file_name = "./exercise2writeto.csv"

my_list = list([])

def csv_to_list(imported_file, my_list):
    with open(imported_file, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            my_list.append(row)

csv_to_list(file_name, my_list)
my_list
```
